```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
```

```python
print(df.head(5))
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
product_freq = df['Product Name'].value_counts()
print(product_freq.head())
```

```python
top_5_stores = df.groupby('Product Name')['Sales'].sum().nlargest(5)
top_5_stores.plot(kind='bar', title='Top 5 Productos con Mayores Ventas')
plt.ylabel('Ventas Totales')
plt.show()
```

```python
plt.hist(df['Segment'], bins=20, color='skyblue')
plt.title('Distribución de los Segmentos')
plt.xlabel('Ventas')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.bar(df['Category'], df['Sales'], alpha=0.5)
plt.title('Relación entre Categoría y Cantidad Vendida')
plt.xlabel('Categoría')
plt.ylabel('Cantidad Vendida')
plt.show()
```
